```python
import numpy as np
import pandas as pd
import seaborn as sns
data = pd.read_csv(r"...\train.csv")
data1 = sns.load_dataset("titanic")
```

```python
data.head()
```

```python
data1.head()
```

```python
data['Embarked'] = data1['embark_town']
```

```python
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.isna().sum()/len(data)*100
```

```python
data.drop(labels='Cabin',axis=1,inplace=True)
data.head()
```

```python
data.duplicated().value_counts()
```

```python
data.Sex.replace(to_replace=['male','female'],value=['Male','Female'],inplace=True)
data.head()
```

```python
data.Embarked.fillna(data.Embarked.mode()[0],inplace=True)
```

```python
data.isna().sum()
```

```python
sns.boxplot(data.Pclass,data.Age)
```

```python
x= data.groupby(by='Pclass',axis=0).get_group(1)['Age'].fillna(round(data.groupby(by='Pclass',axis=0).get_group(1)['Age'].mean()))
y= data.groupby(by='Pclass',axis=0).get_group(2)['Age'].fillna(data.groupby(by='Pclass',axis=0).get_group(2)['Age'].mode()[0])
z= data.groupby(by='Pclass',axis=0).get_group(3)['Age'].fillna(data.groupby(by='Pclass',axis=0).get_group(3)['Age'].mode()[0])
```

```python
data['Age'].loc[x.index] = x
data['Age'].loc[y.index] = y
data['Age'].loc[z.index] = z
```

```python
data.isna().sum().sum()
```

```python
data.to_csv(r'...\Clean_Titanic.csv',index=False)
```
